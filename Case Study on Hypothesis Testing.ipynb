A company started to invest in digital marketing as a new way of their product
promotions. For that they collected data and decided to carry out a study on it.

● The company wishes to clarify whether there is any increase in sales after
stepping into digital marketing.

● The company needs to check whether there is any dependency between

Help the company to carry out their study with the help of data provided.

```python
sales=pd.read_csv(r'/Users/gangashaji/Downloads/Sales_add.csv')
```

```python
sales
```

```python
sales.info()
```

Null Hypothesis:There is no increase in sales after stepping into digital marketing.

Alternate Hypothesis:There is an increase in sales after stepping into digital marketing.

We can use one way anova.

```python
from scipy.stats import f_oneway
```

```python
f_oneway(sales['Sales_before_digital_add(in $)'],sales['Sales_After_digital_add(in $)'])
```

```python
f_stat,p_val=f_oneway(sales['Sales_before_digital_add(in $)'],sales['Sales_After_digital_add(in $)'])
```

```python
p_val
```

```python
if p_val<0.05:
    print('Reject Null Hypothesis')
else:
    print('Accept Null hypothesis')
```

That means there is an increase in sales after stepping into digital marketing.

Part 2

H1:Region doesnot have an impact on the sales

H2:Manager doesnot have an impact on the sales

H3:Region and Manager doesnot have a combined impact on sales

We can use 2 way Anova

```python
sales_melt=pd.melt(sales,id_vars=['Region','Manager'],value_vars=['Sales_before_digital_add(in $)','Sales_After_digital_add(in $)'])
```

```python
sales_melt
```

```python
sales_melt.columns=['Region','Manager','Before/After Digital_Marketing','Value']
```

```python
sales_melt
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model=ols('Value~C(Region)+C(Manager)+C(Region):C(Manager)',data=sales_melt).fit()
```

```python
anova_table=sm.stats.anova_lm(model,typ=2)
```

```python
anova_table
```

Accept H1, H2 and H3 as P is greater than 0.05 for H1 ,H2 and H3

That means region and mananger and they combined doesn't have any impact on the sales.
