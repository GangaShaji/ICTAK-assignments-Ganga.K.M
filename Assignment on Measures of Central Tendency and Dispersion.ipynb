4. What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course

```python
st3=student.groupby(by='gender').std()
```

```python
st3
```

The mean math score for female students is 63.38. Since the standard deviation is 16.02 for math score in females, it indicates that,on an average the math score for female students has a variation of 16.02 marks from the mean math score. As the standard deviation is high,the math scores are relatively inconsistent. But for male students the standard deviation is 14.55 to the mean math score(68.9). This means male students have a lesser score variation from the mean marks,in the math score when compared to female students.
Similar observation can be made regarding the other scores.


```python
st5=student[student['gender']=='female']
st5
```

```python
st6=student[student['gender']=='male']
st6
```

```python
plt.hist(st5['math score'],color='green',edgecolor='black')
plt.axvline(st5['math score'].mean(),linewidth=2)
plt.axvline(st5['math score'].mean()-st5['math score'].std(),linewidth=2,linestyle='dashed')
plt.axvline(st5['math score'].mean()+st5['math score'].std(),linewidth=2,linestyle='dashed')
plt.title('Histogram for math score of female students')
plt.xlabel('math score')
plt.ylabel('frequency')
```

The blue line shows the mean math score for female students and the dashed blue line shows mean-standard deviation.

```python
sns.distplot(st6['math score'],hist=False,color='purple')
plt.axvline(st6['math score'].mean(),linewidth=2,color='green')
plt.axvline(st6['math score'].mean()-st6['math score'].std(),linewidth=2,color='green',linestyle='dashed')
plt.axvline(st6['math score'].mean()+st6['math score'].std(),linewidth=2,color='green',linestyle='dashed')
plt.title('Density plot for math score of male students')
```

The green line shows the mean math score for male students and the dashed green line shows mean-standard deviation.

```python
sns.boxplot([st5['reading score'],st6['reading score']])
plt.ylabel('reading score')
plt.title('Boxplot showing the reading score for female and male students')
```

```python
plt.violinplot([st5['writing score'],st6['writing score']])
plt.ylabel('writing score')
plt.title('Violinplot showing the writing score for female and male students')
```

```python
st4=student.groupby(by='test preparation course').std()
st4
```

The mean math score for students who have completed the test preparation course is 69.9. Since the standard deviation is 14.5 for math score , it indicates that,on an average the math score for students who have completed the test preparation course has a score variation of 14.02 marks from the mean math score. 

When it comes to the standard deviation of math score for students who haven't completed the preparation course, it is 15.75 from the mean score which is 63.8. So the score variation in math score is more in students who haven't done any test prep course.Similar observation can be made regarding the reading and writing scores.


```python
st8=student[student['test preparation course']=='completed']
st8
```

```python
st9=student[student['test preparation course']=='none']
st9
```

```python
sns.boxplot([st8['math score'],st8['reading score'],st8['writing score']])
plt.ylabel('score')
plt.title('Boxplot showing the math,reading and writing scores for students who have completed the test preparation course')
```

```python
sns.boxplot([st9['math score'],st9['reading score'],st9['writing score']])
plt.ylabel('score')
plt.title('Boxplot showing the math,reading and writing scores for students who have not done any test preparation course')
```

So we can either use the measures of central tendency like mean, standard deviation or we can use boxplots or violin plots which is equivalent to quartile measure of dispersion to analyse the scores of different students and their distribution and variation.
